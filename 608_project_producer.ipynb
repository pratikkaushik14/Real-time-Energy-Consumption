```python
from confluent_kafka import Producer
import json
import time
conf = {'bootstrap.servers': '',
         'security.protocol': ',
         'sasl.mechanism': '',
         'sasl.username': '',
         'sasl.password': '',
         'client.id': ""}

kafka_data = []

def chunk_generator(data, chunk_size):
    for i in range(0, len(data), chunk_size):
        yield data[i:i+chunk_size]

def delivery_callback(err, msg):
    if err:
        print('ERROR: Message failed delivery: {}'.format(err))
    else:
        key = msg.key().decode('utf-8')
        kafka_data.append(f"Produced event to kafka topic: key = {key} ")

def save_poll_data_to_file(data_list, file_path):
    with open(file_path, 'w') as file:
        for data in data_list:
            file.write(str(data))
            file.write('\n')  
            
def produce_invoices(json_data, producer):
    counter = 1
    chunk_size = 1000
    for data in chunk_generator(json_data, chunk_size):
        for obj in data:
            time_id = obj["time"]
            producer.produce(topic, key=time_id, value=json.dumps(obj), callback=delivery_callback)
        pool_data = producer.poll(1) 
        time.sleep(1)
        print(str(counter) + " batch completed")
        counter = counter+1
        
            
topic = "topic_2"
kafka_producer = Producer(conf)
produce_invoices(json_data, kafka_producer)
kafka_producer.flush()
save_poll_data_to_file(kafka_data, 'kafka_data.txt')
```
